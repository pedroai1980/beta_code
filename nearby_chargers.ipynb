````python
#-*- coding: utf-8 -*-
import sys
import json
import requests
from geopy.distance import geodesic

# coords are in format [lat,lon]
start    = [59.35 , 18   ]
destin   = [57.755, 14.13]
distance = 150*0.65  # battery autonomy - count margin + roads are not straight
country  = "SE" # ISO country code

transport = "car"
mode      = ["fastest", "shortest", "balanced"][0]
traffic   = "enabled"

response = requests.get("https://api.openchargemap.io/v3/poi/?output=json"+
                        "&countrycode={0}".format(country)+
                        "&latitude={0}&longitude={1}".format(start[0], start[1])+
                        "&distance={0}&distanceunit=KM".format(distance),
                        "&maxresults=500").json()
````

````python
chargers = []
for charger in response:
    # retrieve coordinates from each charging point
    info = charger["AddressInfo"]
    coords = [info["Latitude"], info["Longitude"]]
    # add to list
    chargers.append(coords)

print("Found", len(chargers), "chargers nearby")
````

````python
# expand a charger - returns 5 closest-to-destin chargers
def expand_charger(coords, goal_coords=destin, country=country, distane=distance):
    # get chargers from the API
    response = requests.get("https://api.openchargemap.io/v3/poi/?output=json"+
                            "&countrycode={0}".format(country)+
                            "&latitude={0}&longitude={1}".format(coords[0], coords[1])+
                            "&distance={0}&distanceunit=KM".format(distance),
                            "&maxresults=500").json()
    # get coords
    chargers = []
    for charger in response:
        # retrieve coordinates from each charging point
        info = charger["AddressInfo"]
        coords = [info["Latitude"], info["Longitude"]]
        # add to list
        chargers.append(coords)
        
    # sort by distance to goal. Return 5 closest.
    chargers = [charger for charger in sorted(chargers, key=lambda x: geodesic(x, goal_coords))][:5]
    
    return chargers
````
